### Задание 2
Почему прямые не кажутся перпендикулярными?

```python
x = np.linspace(-5, 5, 21)
y = 3*x + 1
y2 = -1/3 *x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Прямые не кажутся перпендикулярными на данном графике, так как взят некорректный масштаб. Если скорректировать масштаб координатных осей, то станет видно, что прямые перпендикулярны. 

```python
x = np.linspace(-10, 10, 21)
y = 3*x + 1
y2 = -1/3 *x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('scaled')
plt.show()
```

### Задание 3

Напишите код на Python, реализующий построение графиков:
окружности,	
эллипса,
гиперболы.

```python
#уравнение окружности: x**2 + y**2 = r**2.
x1= []
x2= []
y1 = []
y2 = []
r = 100

for i in range(10000):
    x1.append(i)
    x2.append(-i)
    y1.append(np.sqrt(r**2 - i**2))
    y2.append(- np.sqrt(r**2 - i**2))

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.axis('scaled')
plt.show()
```

```python
# Уравнение эллипса x**2/a**2 + y**2/b**2 = 1

a = 50
b = 30
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(1000):
    x1.append(i)
    x2.append(-i)
    y1.append(np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.axis('scaled')
plt.show()
```

```python
# Уравнение гиперболы x**2/a**2 - y**2/b**2 = 1
a = 50
b = 30
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(1000):
    x1.append(i)
    x2.append(-i)
    y1.append(np.sqrt(b ** 2 + (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 + (b ** 2 * (i ** 2 / a ** 2))))

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.axis('scaled')
plt.show()
```

x = R*cos(a)
y = R*sin(a)
R = np.sqrt(x**2 + y**2)

```python
import math
```

```python
def polar_to_decart(R, a):
    x = R*math.cos(a)
    y = R*math.sin(a)
    return x, y

polar_to_decart(5, 45)
```

###  Задание (в программе)
exp(x) + x∙(1 – y) = 1  
y = x2 – 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-2, 3, 201)
y1 = (np.exp(x)+ x - 1)/x
y2 = x**2 - 1

plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.xlim(-3, 4)
plt.ylim(-3, 9)
plt.grid(True)

plt.legend()
plt.show()
```

```python
def equations(p):
    x, y = p
    return (x**2- y - 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (2, 4))
x2, y2 = fsolve(equations, (-1, 1))

print(x1, y1)
print(x2, y2)
```
